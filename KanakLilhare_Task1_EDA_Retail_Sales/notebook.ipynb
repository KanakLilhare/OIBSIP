# Data Loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
sns.set_theme()

# Display all columns
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv(
    "data/Supermart Grocery Sales - Retail Analytics Dataset.csv"
)

df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.head()
```

# Data Cleaning

```python
# Check missing values
df.isnull().sum()

# Check duplicate rows
df.duplicated().sum()

# Remove duplicates if any
df = df.drop_duplicates()
```

```python
df['Order Date'].head(10)
```

```python
df['Order Date'].unique()[:10]
```

```python
df['Order Date'] = pd.to_datetime(
    df['Order Date'],
    format='mixed'
)
```

```python
df['Order Date'].head()
```

# Descriptive Statistics

```python
df[['Sales','Profit','Discount']].describe()
```

```python
print("Mean Sales:", df['Sales'].mean())

print("Median Sales:", df['Sales'].median())

print("Mode Sales:")
print(df['Sales'].mode())

print("Standard Deviation of Sales:", df['Sales'].std())
```

# Time Series Analysis

```python
df['Month_Year'] = df['Order Date'].dt.to_period('M')
```

```python
df[['Order Date', 'Month_Year']].head()
```

```python
monthly_sales = df.groupby('Month_Year')['Sales'].sum()
monthly_sales.head()
```

```python
plt.figure(figsize=(14,6))

monthly_sales.plot(marker='o')

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')

plt.grid(True)

plt.show()
```

```python
monthly_sales.sort_values(
    ascending=False
).head(10)
```

# Product Analysis

```python
category_sales = df.groupby(
    'Category'
)['Sales'].sum()

category_sales
```

```python
category_sales = df.groupby('Category')['Sales'].sum()

plt.figure(figsize=(12,6))

sns.barplot(
    x=category_sales.index,
    y=category_sales.values
)

plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')

plt.xticks(rotation=30, ha='right')

plt.tight_layout()

plt.show()
```

# Sub Category Analysis

```python
subcategory_sales = df.groupby(
    'Sub Category'
)['Sales'].sum().sort_values(
    ascending=False
)

subcategory_sales.head(10)
```

```python
plt.figure(figsize=(12,6))

subcategory_sales.head(10).plot(
    kind='bar'
)

plt.title('Top 10 Sub Categories')
plt.xlabel('Sub Category')
plt.ylabel('Sales')

plt.show()
```

# Customer Analysis

```python
top_customers = df.groupby(
    'Customer Name'
)['Sales'].sum().sort_values(
    ascending=False
).head(10)

top_customers
```

```python
plt.figure(figsize=(12,5))

top_customers.plot(kind='bar')

plt.title('Top 10 Customers')

plt.ylabel('Sales')

plt.show()
```

# Visualizations

```python
region_sales = df.groupby(
    'Region'
)['Sales'].sum()

region_sales
```

```python

region_sales = df.groupby('Region')['Sales'].sum()

plt.figure(figsize=(8,8))

plt.pie(
    region_sales,
    labels=region_sales.index,
    autopct='%1.1f%%',
    wedgeprops={'width':0.4}
)

plt.title("Regional Sales Contribution")

plt.show()
```

```python
top_states = df.groupby(
    'State'
)['Sales'].sum().sort_values(
    ascending=False
).head(10)

top_states

plt.figure(figsize=(10,5))

top_states.plot(kind='bar')

plt.title('Top 10 States by Sales')
plt.xlabel('State')
plt.ylabel('Sales')

plt.show()
```

```python
category_profit = df.groupby('Category').agg({
    'Sales':'sum',
    'Profit':'sum'
}).reset_index()

plt.figure(figsize=(10,6))

sns.scatterplot(
    data=category_profit,
    x='Sales',
    y='Profit',
    s=300
)

for i in range(len(category_profit)):
    plt.text(
        category_profit['Sales'][i],
        category_profit['Profit'][i],
        category_profit['Category'][i]
    )

plt.title('Category-wise Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')

plt.show()
```

```python
monthly_sales = df.groupby('Month_Year')['Sales'].sum()

plt.figure(figsize=(15,6))

plt.plot(
    monthly_sales.index.astype(str),
    monthly_sales.values,
    marker='o'
)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.xticks(rotation=45)

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8,5))

corr = df[['Sales','Profit','Discount']].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap='RdYlBu',
    linewidths=1
)

plt.title("Correlation Matrix")

plt.show()
```

```python
category_sales = df.groupby('Category')['Sales'].sum()

plt.figure(figsize=(8,8))

plt.pie(
    category_sales,
    labels=category_sales.index,
    autopct='%1.1f%%'
)

plt.title("Sales Share by Category")

plt.show()
```

# Insights

1. Eggs, Meat & Fish generated the highest sales revenue among all product categories.

2. Oil & Masala recorded the lowest sales and may require promotional efforts.

3. Sales demonstrate monthly fluctuations, indicating seasonal demand patterns.

4. A small group of customers contributes significantly to overall revenue.

5. High-performing regions and states can be targeted for expansion strategies.

6. Correlation analysis helps understand the relationship between sales, profit, and discounts.

# Recommendations

1. Increase inventory availability for Eggs, Meat & Fish and Snacks categories because they contribute the highest sales.

2. Introduce promotional campaigns for Oil & Masala products to improve sales performance.

3. Implement a customer loyalty program targeting top customers such as Krithika, Amrish, and Verma.

4. Use seasonal sales trends to optimize inventory planning and avoid stock shortages.

5. Analyze products with high discounts but low profit margins to improve profitability.

6. Focus marketing efforts on high-performing categories while developing strategies for lower-performing categories.

# Conclusion

The exploratory data analysis of the Supermart Grocery Sales dataset revealed valuable insights into sales performance, customer behavior, and product categories.

Eggs, Meat & Fish emerged as the highest-performing category, while Oil & Masala generated comparatively lower sales. Customer analysis showed that a small number of customers contribute significantly to revenue generation.

The findings can help management make informed decisions regarding inventory management, customer retention, promotional campaigns, and overall business strategy.

Data visualization and statistical analysis provided a clear understanding of sales patterns and business performance, enabling data-driven decision-making.
